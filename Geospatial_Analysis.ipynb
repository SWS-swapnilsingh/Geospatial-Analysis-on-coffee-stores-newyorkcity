```python
import pandas as pd
import numpy as np
import geopy.distance
import folium #helpful to visualize map
```

```python
df = pd.read_csv('https://drive.google.com/uc?id=19JERBe0ndadkMPKDFZ0rTEZqdI50zDfk')
```

```python
df.head()
```

```python
df['Coordinates'] = '(' + df['lat'].astype(str) + ',' + df['lng'].astype(str) + ')'
df.head(5)
```

```python
ESB = [40.74855426290525, -73.98493484348859]  # ESB - Empire State Building
m = folium.Map(location = ESB, zoom_start=12)

for i, row in df.iterrows():
  lat = df.loc[i, 'lat']
  lng = df.loc[i, 'lng']
  store = df.loc[i, 'store']

  if store == 'Starbucks':
    color = 'green'
  else:
    color = 'red'

  folium.Marker(location = [lat, lng], popup=store, icon = folium.Icon(color = color)).add_to(m)

m
```

```python
distance = []
for row in df.itertuples():
  coordinate_str = row.Coordinates
  # Remove the parentheses and split the string by the comma
  coordinate_parts = coordinate_str.strip('()').split(',')
  # Convert the parts to floats and create a tuple
  coordinate_tuple = (float(coordinate_parts[0]), float(coordinate_parts[1]))

  distance.append(geopy.distance.distance(ESB, coordinate_tuple).km)

df['Distance'] = distance
df.head(5)
```

```python
#Number of 'Starbucks' stores that are with 5 km range from Empire State Building
df[(df['Distance'] <= 5) & (df['store'] == 'Starbucks')].shape[0]
```

```python
df['store'].unique()
```

```python
#Number of 'McDonalds' stores that are with 5 km range from Empire State Building
df[(df['Distance'] <= 5) & (df['store'] == 'McDonalds')].shape[0]
```

```python
print(f'Starbucks stores are {203/39:.0f} times more than McDonals stores, in the 5 km range of Empire State Building')
```

```python
#Mean, Max, and Min distance of 'Starbucks' and 'McDonalds' store from Empire State Building
df.groupby('store')['Distance'].agg(['mean', 'max', 'min'])
```

so, on an average 'McDonald' stores are at greater distance than 'Starbucks' stores from Empire State Building
